```python
df=pd.read_csv("E-commerce Dataset.csv")

df.head()
```

```python
# Basic structure and summary
df.info()
df.describe(include='number').T
```

```python
print("Gender counts:")
print(df['gender'].value_counts(), "\n")
```

```python
print("Country counts:")
print(df['country'].value_counts().head(10), "\n")

print("Product categories:")
print(df['category'].value_counts(), "\n")

print("Payment methods:")
print(df['payment_method'].value_counts(), "\n")
```

```python
df['total_amount'] = df['quantity'] * df['unit_price']
df.head()
```

```python
print("Missing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
num_cols=['quantity','unit_price','rating','total_amount']

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=10, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in['unit_price','total_amount']:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='category', data=df, order=df['category'].value_counts().index)
plt.title("Number of orders per category")
plt.xlabel("count")
plt.ylabel("category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='rating',y='category', data=df, estimator='mean')
plt.title("Average Rating by product category")
plt.xlabel("Average Rating")
plt.ylabel("category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['quantity','unit_price','rating','total_amount']].corr(), annot=True, cmap='coolwarm')

plt.title("Correlation between Numeric variables")
plt.show()
```

```python
sns.pairplot(df[['quantity','unit_price','rating','total_amount']])
plt.show()
```

```python
## Summary of Findings
- Dataset contains 119 transactions and 15 features.
- No missing or duplicate data.
- Electronics and Apparel dominates sales.
- Cash on Delivery is the most common payment method.
- Ratings mostly range 2-5 with average around 3.
- Higher quantity and unit_price lead to higher total_amount.
- A few outliers exist in total_amount (bulk orders).
```
